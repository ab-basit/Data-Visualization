```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

## Grids
* **Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.**

## PairGrid
* **Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.**

```python
sns.PairGrid(iris)
```

```python
grid=sns.PairGrid(iris)
grid.map(plt.scatter)##grid.map(plt.plot)
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_lower(plt.scatter)
g.map_upper(sns.kdeplot)
```

## pairplot 
* **pairplot is the simple version of pairgrid**

```python
sns.pairplot(iris)
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips=sns.load_dataset('tips')
```

```python
sns.FacetGrid(tips,row='sex',col='smoker')
```

```python
g1=sns.FacetGrid(tips,row='sex',col='smoker')
g1.map(plt.hist,'tip')
```

```python
g2=sns.FacetGrid(tips,row='sex',col='smoker',hue='time')
g2.map(plt.scatter,'total_bill','tip').add_legend()
```

## JointGrid

JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g3=sns.JointGrid(x='total_bill',y='tip',data=tips)
g3.plot(sns.regplot,sns.distplot)
```
